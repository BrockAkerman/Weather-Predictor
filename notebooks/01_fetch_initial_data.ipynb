```python
# ⚠️ BOOTSTRAP NOTEBOOK — DO NOT RUN AGAIN
# This notebook is only for first-time initialization.
# It should not be re-executed in normal operation.

import requests
import json
import glob
import sys
from pathlib import Path
from datetime import datetime

# -----------------------------------------
# SAFEGUARD: Abort if bronze already has data
# -----------------------------------------
existing_files = glob.glob("../data/bronze/raw_*.json")

if existing_files:
    print("❌ Bronze layer already contains raw files. "
          "This bootstrap notebook should NOT be re-run.")
    print("Existing files:")
    for f in existing_files:
        print("  -", f)
    sys.exit("Aborting to avoid overwriting initial bootstrap data.")

# -----------------------------------------
# 1. Your Open-Meteo endpoint
# -----------------------------------------
url = "https://api.open-meteo.com/v1/forecast?latitude=34.172&longitude=-83.5588&daily=uv_index_max,shortwave_radiation_sum&hourly=temperature_2m,relative_humidity_2m,dew_point_2m,precipitation_probability,precipitation,cloud_cover,surface_pressure,wind_speed_10m,wind_gusts_10m,wind_direction_10m&timezone=America/New_York&past_days=1&forecast_days=7&wind_speed_unit=ms"

# -----------------------------------------
# 2. Fetch JSON data
# -----------------------------------------
response = requests.get(url)
data = response.json()

# -----------------------------------------
# 3. Ensure bronze directory exists
# -----------------------------------------
bronze_path = Path("../data/bronze")
bronze_path.mkdir(parents=True, exist_ok=True)

# -----------------------------------------
# 4. Save with timestamp
# -----------------------------------------
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
output_file = bronze_path / f"raw_{timestamp}.json"

with open(output_file, "w") as f:
    json.dump(data, f, indent=2)

output_file
```
